# Identification and Removal of Outliers

```python
#boxplot of dataset
sns.boxplot(df2)
```

```python
Q1 = df2['Sentiment_Polarity'].quantile(0.25)
Q3 = df2['Sentiment_Polarity'].quantile(0.75)
IQR = Q3-Q1
threshold=1.5 
admit_outliers=df2[(df2['Sentiment_Polarity'] <Q1-threshold*IQR) | (df2['Sentiment_Polarity'] > Q3+ threshold*IQR)]
admit_outliers
```

```python
df2=df2.drop(admit_outliers.index)
sns.boxplot(x='Sentiment_Polarity',data=df2)
```

```python
Q1 = df2['Sentiment_Subjectivity'].quantile(0.25)
Q3 = df2['Sentiment_Subjectivity'].quantile(0.75)
IQR = Q3-Q1
threshold=1.5 
admit_outliers=df2[(df2['Sentiment_Subjectivity'] <Q1-threshold*IQR) | (df2['Sentiment_Subjectivity'] > Q3+ threshold*IQR)]
admit_outliers
```

```python
df2=df2.drop(admit_outliers.index)
sns.boxplot(x='Sentiment_Subjectivity',data=df2)
```

# Machine Learning

```python
# Required libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

```python
# Feature Selection
X = df[['Rating', 'Reviews', 'Sentiment_Polarity', 'Sentiment_Subjectivity']].fillna(0)
y = df['Sentiment']
```

```python
#cEncode Target Variable
le = LabelEncoder()
y_encoded = le.fit_transform(y)  
```

```python
#Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

```python
# Feature Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Define models
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'KNN': KNeighborsClassifier(),
    'Naive Bayes': GaussianNB(),
    'SVM': SVC()
}
```

```python
#Train & Evaluate
results = {}
for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    accuracy = accuracy_score(y_test, y_pred)
    results[name] = accuracy*100
```

```python
print("Model Accuracy Comparison:")
for model_name, acc in results.items():
    print(f"{model_name}: {acc:.4f}")
```

```python
model_names = list(results.keys())
accuracies = list(results.values())

# Plotting the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=model_names, y=accuracies, palette='viridis')
plt.ylabel("Accuracy")
plt.ylim(10, 100)
plt.title("Model Accuracy Comparison")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
```

```python
Based on the model accuracy comparison, the following conclusions can be drawn:

1. Decision Tree and Random Forest exhibit outstanding performance with accuracy values nearing 100%, demonstrating their effectiveness 
in handling the dataset and capturing underlying patterns. These models are particularly strong in this scenario, likely due to their ability
to deal with complex, non-linear relationships.

2. KNN, Naive Bayes, and SVM also perform well with accuracies in the high 90s, indicating that these models are robust and able to generalize 
effectively on this data.

3.Logistic Regression, while still performing well, has a comparatively lower accuracy of 88.42%. This suggests that it might not be as
effective for the complexities of this dataset, possibly due to its linear nature and inability to capture more intricate patterns compared to 
tree-based models.
```
